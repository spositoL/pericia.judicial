```python
data = pd.read_csv('respostas.csv')
data1 = pd.read_csv('respostas.csv')
data.head(1)
```

```python
#vamos renomear o nome de totdas as colunas para C1, C2, C3, etc
data.columns = [f'C{i}' for i in range(1, len(data.columns) + 1)]
data.head()
```

```python
#Quero analisar a coluna C4 de forma a separar palavras chaves e eliminar palavras irrelevantes. 
# Para isso, vou usar a biblioteca nltk para processamento de linguagem natural.
#%pip install nltk
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('portuguese'))
```

```python
# Gerar matriz com indice e nome das colunas de data1 e data
colunas = pd.concat([
    pd.Series(range(len(data1.columns)), name='index'),
    pd.Series(data1.columns, name='nome_coluna'),
    pd.Series(range(len(data.columns)), name='index_final'),
    pd.Series(data.columns, name='nome_coluna_final')
] , axis=1)

colunas
```

```python
import re
import numpy as np
import seaborn as sns
from collections import Counter

questoes = ['C4', 'C10', 'C16', 'C22', 'C25', 'C31']
questoes_validas = [c for c in questoes if c in data.columns]
questoes_ausentes = [c for c in questoes if c not in data.columns]

if questoes_ausentes:
    print(f'Colunas ausentes e ignoradas: {questoes_ausentes}')

# Stopwords adicionais comuns em respostas abertas
stopwords_extras = {
    'pra', 'pro', 'q', 'né', 'nao', 'não', 'sim', 'tambem', 'também',
    'muito', 'coisa', 'coisas', 'ser', 'estar', 'ter', 'sobre', 'pode', 'pois'
}
stop_words_total = stop_words.union(stopwords_extras)

def limpar_tokenizar(texto):
    texto = texto.lower()
    # Mantem apenas letras (incluindo acentos) e espacos
    texto = re.sub(r"[^a-zà-ÿ\s]", " ", texto)
    return [t for t in texto.split() if len(t) > 2 and t not in stop_words_total]

registros_kw = []
tokens_por_questao = {}

for col in questoes_validas:
    textos = data[col].fillna('').astype(str)
    tokens_series = textos.apply(limpar_tokenizar)
    data[f'{col}_tokens'] = tokens_series
    tokens_por_questao[col] = tokens_series

    freq_col = Counter([palavra for lista in tokens_series for palavra in lista])
    for palavra, frequencia in freq_col.items():
        registros_kw.append({
            'questao': col,
            'palavra': palavra,
            'frequencia': frequencia
        })

data_kw = pd.DataFrame(registros_kw).sort_values(
    ['questao', 'frequencia'], ascending=[True, False]
).reset_index(drop=True)

print('DataGrid data_kw criado com sucesso.')
display(data_kw.head(30))

# Top 15 por questao para inspeção rápida
top_por_questao = (
    data_kw.sort_values(['questao', 'frequencia'], ascending=[True, False])
    .groupby('questao')
    .head(15)
    .reset_index(drop=True)
)
display(top_por_questao)
```

```python
# Visualizacoes solicitadas a partir do data_kw
from wordcloud import WordCloud

# Palavras globais (independente das questoes)
freq_global = (
    data_kw.groupby('palavra', as_index=False)['frequencia']
    .sum()
    .sort_values('frequencia', ascending=False)
)

# Define o recorte de keywords para os heatmaps
top_n = 20
top_keywords = freq_global.head(top_n)['palavra'].tolist()

# (a) Nuvem de palavras global
wordcloud = WordCloud(
    width=1800,
    height=1000,
    background_color='white',
    colormap='viridis',
    max_words=150,
    collocations=False
).generate_from_frequencies(dict(zip(freq_global['palavra'], freq_global['frequencia'])))

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nuvem de Palavras Global (C4, C10, C16, C22, C25, C31)', fontsize=15)
plt.show()

# Monta matriz binaria por resposta para coocorrencia palavra x palavra
tokens_por_resposta = []
for idx in data.index:
    tokens_linha = []
    for col in questoes_validas:
        tokens_linha.extend(data.at[idx, f'{col}_tokens'])
    tokens_por_resposta.append(set(tokens_linha))

mat_bin = pd.DataFrame(0, index=data.index, columns=top_keywords, dtype=int)
for i, tokens_set in enumerate(tokens_por_resposta):
    presentes = [kw for kw in top_keywords if kw in tokens_set]
    if presentes:
        mat_bin.loc[i, presentes] = 1

# (b) Heatmap cruzando palavras-chave identificadas (coocorrencia palavra x palavra)
cooc_kw = mat_bin.T.dot(mat_bin)
cooc_arr = cooc_kw.to_numpy(copy=True)
np.fill_diagonal(cooc_arr, 0)
cooc_kw = pd.DataFrame(cooc_arr, index=cooc_kw.index, columns=cooc_kw.columns)

plt.figure(figsize=(12, 10))
sns.heatmap(cooc_kw, cmap='YlGnBu', linewidths=0.3)
plt.title('Heatmap de Coocorrencia entre Palavras-chave', fontsize=14)
plt.xlabel('Palavras-chave')
plt.ylabel('Palavras-chave')
plt.tight_layout()
plt.show()

# (c) Heatmap cruzando questoes com palavras-chave
qk_matrix = (
    data_kw[data_kw['palavra'].isin(top_keywords)]
    .pivot_table(index='questao', columns='palavra', values='frequencia', aggfunc='sum', fill_value=0)
    .reindex(index=questoes_validas, columns=top_keywords, fill_value=0)
)

plt.figure(figsize=(14, 6))
sns.heatmap(qk_matrix, cmap='OrRd', annot=True, fmt='d', linewidths=0.3)
plt.title('Heatmap Questoes x Palavras-chave', fontsize=14)
plt.xlabel('Palavras-chave')
plt.ylabel('Questoes')
plt.tight_layout()
plt.show()

display(freq_global.head(30))
```

## Insight final da análise da C31

A análise da C31 indica com clareza que a mudança de maior impacto para converter visitante em cliente é **destacar de forma imediata o caminho de contato com o perito** (termos recorrentes: *contato*, *cliente*, *visitante*, *destacar*, *solicitar*). À luz de Evolução, Usabilidade, Heurísticas e Ergonomia, isso significa reduzir ambiguidade no fluxo e transformar a ação principal em algo visível, previsível e de baixo esforço cognitivo desde a primeira dobra da página.

Em termos práticos, a recomendação central é implementar um **CTA primário e persistente** (ex.: "Falar com o Perito") com alto contraste, posicionamento estratégico e reforço por microtextos de confiança. Essa intervenção concentra o ganho porque melhora encontrabilidade, reforça affordance de ação e reduz a fricção de navegação; assim, o usuário deixa de explorar sem direção e passa a seguir um percurso claro até a conversão.

## Parecer final por questão (análise descritiva)

### C4 - Encontrabilidade e primeira impressão
Os resultados da C4 indicam que a percepção de qualidade começa na capacidade de o usuário encontrar rapidamente o site e compreender sua proposta logo no primeiro contato. A recorrência de termos ligados a "encontrabilidade", "site", "usuário" e "interface" sugere que a visibilidade inicial no fluxo de busca e a clareza da página de entrada funcionam como gatilhos de confiança. Em termos de ergonomia cognitiva, o sistema é melhor avaliado quando reduz incerteza desde os primeiros segundos de navegação.

### C10 - Organização da informação e carga cognitiva
Na C10, os participantes destacam "chunking", "blocos", "agrupamento" e referências à regra de Miller, evidenciando compreensão de que a estrutura do conteúdo influencia diretamente a compreensão. O parecer é que a organização em unidades menores facilita leitura, memória de trabalho e tomada de decisão, reduzindo sobrecarga mental. Trata-se de um ponto central de ergonomia cognitiva: quanto mais previsível e segmentada a arquitetura informacional, maior a fluidez da interação.

### C16 - Legibilidade e conforto visual
A C16 concentra termos como "leitura", "espaçamento", "linhas", "visual" e "contraste", apontando que o conforto visual é decisivo para retenção e continuidade da navegação. O diagnóstico sugere que melhorias tipográficas (entrelinha, contraste, hierarquia textual e comprimento de linha) têm alto impacto na percepção de usabilidade. Aqui há convergência entre ergonomia física (esforço ocular, fadiga) e ergonomia cognitiva (facilidade de processamento da informação).

### C22 - Efetividade percebida e confiança no conteúdo
Na C22, aparecem com força "site", "profissional", "judicial", "eficácia", "satisfação" e "análise", indicando que os usuários associam credibilidade do serviço à clareza da interface e à coerência do conteúdo. O parecer é que a experiência é percebida como funcional, porém com espaço para aprimorar sinais de orientação e continuidade de ação. Em termos de design, confiança e conversão aumentam quando linguagem, estrutura e evidências de autoridade são apresentados de forma simples e progressiva.

### C25 - Affordance de conversão (falar com o perito)
A C25 evidencia que o elemento de affordance mais relevante para conversão é o CTA de contato com o perito (botão/ação de solicitar atendimento). Apesar de parte dos respondentes considerar o fluxo "óbvio", também surgem sinais de fricção que indicam necessidade de busca ativa para localizar o próximo passo. O parecer técnico é reforçar visibilidade, contraste, posição e consistência do CTA para que a intenção de contato seja acionada com menor esforço cognitivo e menor ambiguidade.

### C31 - Mudança de maior impacto para transformar visitante em cliente
A C31 converge para uma recomendação prioritária: tornar o caminho de conversão explícito e persistente ao longo da jornada. Termos como "contato", "cliente", "visitante", "mudança" e "destacar" confirmam que a principal oportunidade está na clareza do percurso de ação. O parecer final da questão é implementar um CTA primário com hierarquia visual forte, reforços de confiança e presença contínua em pontos estratégicos da página.

## Conclusão geral da análise - Ergonomia cognitiva e física no design
No conjunto das questões C4, C10, C16, C22, C25 e C31, os dados mostram que o desempenho do site depende da integração entre **ergonomia cognitiva** (clareza, previsibilidade, organização em blocos, orientação do fluxo) e **ergonomia física** (legibilidade, contraste, espaçamento e redução de fadiga visual). A experiência melhora quando o usuário entende rapidamente onde está, o que deve fazer e como concluir a ação principal sem esforço excessivo.

Como diretriz estratégica única, a intervenção de maior impacto é consolidar um **fluxo de conversão claro e contínuo**, apoiado por arquitetura informacional enxuta, leitura confortável e CTA de contato altamente visível. Essa combinação reduz carga mental, minimiza fricções de navegação e aumenta a probabilidade de transformar visitante em cliente com maior eficiência e confiança.
